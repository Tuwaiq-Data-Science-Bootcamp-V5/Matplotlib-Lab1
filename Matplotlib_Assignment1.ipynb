```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python

df1=[10,20,30]
df = pd.DataFrame({
    'Color' : [
        'Red' if i in range(1, 45+1) else 'Green' if i in range(45+1, 45+70+1) else 'Blue' for i in range(1, 145+1)
    ]
})
df
```

```python
df['Color'].value_counts()
```

```python
categories = list(df['Color'].value_counts().index)
counts     = df1

plt.bar(
    x = categories,            # The x axes
    height = counts,           # The y axes
)

plt.show();
```

```python
plt.bar(range(len(df1)),df1)
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(
    df1, labels=range(len(df1)),
    autopct='%1.1f%%', explode = [0.4,0.4,0.4]
)
plt.show();
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(
    df1,df1,
    color='b', alpha=0.4,
    marker = 'o', s=50
)

plt.show();
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(
    df1,df1,
    color='b', alpha=0.4,
    marker = 'o', s=50
)
plt.title('Pie Chart',loc='left',color='red')
plt.xlabel("X axis",fontsize=16)
plt.ylabel("Y axis",fontsize=16)
plt.show();
```

## Q5: Create three line charts with Legends at left and colors.

```python
width=0.25
plt.bar(np.arange(len(df1))-width,df1,width,color='red')
plt.bar(np.arange(len(df1)),df1,width,color='grey')
plt.bar(np.arange(len(df1))+width,df1,width,color='green')
# plt.xticks(np.arange(len(df1)), list(prod['Products'].value_counts().index))
plt.legend(['r','g','gg'])
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(20, 10))

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

axs[0].plot(x, y1, color='r')
axs[0].set_title('Sin Plot')

axs[1].plot(x, y2, color='b')
axs[1].set_title('Cos Plot')

axs[2].plot(x, y1, color='g')
axs[2].set_title('Sin Plot')
plt.show()
```
