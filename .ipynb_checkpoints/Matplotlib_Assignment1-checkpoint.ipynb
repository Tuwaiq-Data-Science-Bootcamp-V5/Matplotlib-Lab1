```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
lstx = [1,2,3]
lsty = [1,4,9]
df = pd.DataFrame({"age":[20,30,40,50,60],"speed":[20,18,16,14,8],"nationalty":["saudi","saudi","UAE","saudi","USA"]})
```

```python
plt.bar(
    x = lstx,
    height = lsty,
)

plt.show();
```

```python
plt.scatter(
    df["age"],
    df["speed"],     )
plt.show();
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(
    
    df.value_counts("nationalty"),
    explode= [1,1,1]
)

plt.show();
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(
    df["age"],
    df["speed"],
    alpha=0.4,
    s=50,

    )
plt.show();
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(
    df["age"],
    df["speed"],
    alpha=0.4,
    s=50,

    )

plt.title('my scatter plot', loc='left',c='r')
plt.xlabel("xlabel",fontsize = 16)
plt.ylabel("ylabel",fontsize = 16)


plt.show();
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot(df["age"])
plt.plot(df["speed"])
plt.plot(df["nationalty"])
plt.legend(["age","speed","nationalty"],loc="upper left")
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))


axs[0, 0].plot(df["age"], df["speed"])


axs[0, 1].plot(df["nationalty"], df["speed"])

axs[1, 0].plot(df["nationalty"], df["age"])

plt.show();
```
