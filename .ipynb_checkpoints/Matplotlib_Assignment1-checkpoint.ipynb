```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
singer=["Ahlam", "Abdulmajeed", "Rabeh", "Majed", "Oumaima", "Khalid", "Sia", "Imagine dragons", "Adele"]
songs=[45,30,23,10,5,15,18,13,10]

ig = plt.figure(figsize=(12, 6))
plt.bar(
    x = singer,            
    height = songs,         
)

plt.show();
```

```python
df=pd.DataFrame(columns=['Singer','Number of Songs'])
df['Singer'] = singer
df['Number of Songs'] = songs
df

colors = ['pink', 'silver', 'steelblue','lavender','teal','tan'] 
df.groupby(['Singer']).sum().plot( 
    kind='pie', y='Number of Songs',  
  autopct='%1.0f%%', colors=colors) 

```

## Q2: Create pie chart with equal explode values.

```python
y = np.array([45,30,23,10,5,15,18,13,10])
myexplode = [0.2, 0, 0, 0,0,0,0,0,0]
colors = ['pink', 'silver', 'steelblue','lavender','teal','tan'] 

plt.pie(y, labels = singer, explode = myexplode, colors = colors)
plt.show() 
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
ig = plt.figure(figsize=(10, 3))
sizes = np.random.rand(50)

plt.scatter(
    df['Singer'], df['Number of Songs'],
    color='purple', alpha = 0.4, label='Number of Songs'
)

plt.title('Number of songs in MyFav Lists', fontsize=15)
plt.xlabel('Singers', fontsize=12)
plt.ylabel('Number of songs', fontsize=12)
plt.legend()

plt.show();
```

```python
points=[2,4,6,8,10]
playerA= [3,8,1,10,10]
playerB= [5,4,5,10,8]
playerC= [7,3,10,10,10]

x = [1,2,3,4,5] 
y = [3,3,3,3,3] 

plt.plot(points, playerA, label = "playerA", linestyle="--",color='teal') 
plt.plot(points, playerB,label = "playerB", linestyle="--",color='silver') 
plt.plot(points, playerC, label = "playerC", linestyle="--",color='tan') 
plt.legend(labelcolor='black', loc = 'upper left')
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
fig = plt.figure(figsize=(20, 10))

x = np.linspace(0, 1* np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1,3,1)
plt.plot(x, y1, color='purple')
plt.title('Sin Plot')

plt.subplot(1,3,2)
plt.plot(x, y2, color='silver')
plt.title('Cos Plot')

plt.subplot(1,3,3)
plt.plot(x, y1, color='teal')
plt.title('Sin Plot')

plt.show();
```
