```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
lst= ['zayed','ahmed','rayan','ali']
lst2= [12000,9000,6000,4000]
df_data={'accupation':['engineer','pharmacy','health care','tourism'],
         'age':["25","24","23","22"],
         'city':["jeddah", "riyadh", "khobar","abha"],
        'salary': ["12000","9000","6000","4000"]}

df=pd.DataFrame(df_data)
df.describe
```

```python
plt.bar(lst,lst2,color= 'r')
```

```python
plt.hist(df.value_counts("city"))
plt.xlabel('accupation')
plt.ylabel('city')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(df["age"])
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df['age'].value_counts().values, df['salary'].value_counts().values, marker='o', alpha=0.4, s=50)
plt.show()
```

```python
age_value_counts = df['age'].value_counts()
salary_value_counts = df['salary'].value_counts()

plt.scatter(age_value_counts.values, salary_value_counts.values, marker='o', alpha=0.4, s=50)

plt.title("my scatter plot", loc='left', c='r')
plt.xlabel("xlabel", fontsize=16)
plt.ylabel("ylabel", fontsize=16)
```
